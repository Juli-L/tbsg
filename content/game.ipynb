```python
import time
import random

def start_game():
    global has_armor, has_sword, has_shield, has_magic, is_rich, is_poor, inventory, Häuser, konten, name
    has_armor = False
    has_sword = False
    has_shield = False
    has_magic = False
    is_rich = False
    is_poor = True
    inventory = []
    Häuser = []
    konten = {
        "gold": 0 
    }
    name = dragon_name()
   
    print("Willkommen zu deinem Abenteuer!")
    time.sleep(1)
    show_status()
    proceed_to_dragon()


def Kreuzung():    
    print("Du stehst an einer Weggabelung im Wald. Wohin möchtest du gehen?")
    time.sleep(1)
    decision = input("Links oder rechts? (l/r): ").strip().lower()
    
    if decision == "l":
        find_gold()
        River()
    elif decision == "r":
        find_gold()
        acquire_item()
        Markt()
    else:
        print("Ungültige Eingabe. Bitte gib 'l' oder 'r' ein.")
        Kreuzung()

def show_status():
    print(f"Gold: {konten['gold']} ")
    print(f"Inventar: {inventory}")
    print(f"Häuser: {Häuser}")

def explore_area():
    locations = [
        "eine alte Ruine", "ein dunkler Tunnel", "ein verborgener Tempel", "ein verfallenes Schloss",
        "ein dichter Bambuswald", "eine verzauberte Lichtung", "eine verborgene Bibliothek", "ein mystischer Brunnen",
        "ein unterirdisches Verlies", "ein vergessener Friedhof", "eine Kristallhöhle", "ein geheimer Altar",
        "ein verlassener Wachturm", "eine Schmiede voller Waffen", "eine Höhle mit uralten Inschriften",
        "ein Feld mit magischen Blumen", "eine Insel im Nebel", "ein verborgener Handelsposten",
        "ein uraltes Drachenheiligtum", "eine Höhle voller Leuchtpilze"
    ]
    location = random.choice(locations)
    print(f"Du erkundest {location} und findest etwas Nützliches!")
    find_gold()
    acquire_item()
    show_status()

def find_gold():
    found_gold = random.randint(5, 50)
    konten["gold"] += found_gold
    print(f"Du hast {found_gold} Goldmünzen gefunden! Insgesamt besitzt du jetzt {konten['gold']} Gold.")

def acquire_item():
    global has_armor, has_sword, has_shield, has_magic, is_rich, is_poor, inventory
    items = [
        "Magische Rüstung", "Flammenschwert", "Drachenschild", "Heiltrank", "Goldene Pfeile", "Geheimnisvolle Schriftrolle",
        "Lichtamulett", "Zauberstab", "Diamantdolch", "Schattenmantel", "Runenstein", "Eiszauber",
        "Heilige Kette", "Verstärkter Helm", "Blitzspeer", "Tarnumhang", "Dämonenring", "Verzauberte Schuhe",
        "Feueramulette", "Schutzring", "Villa"
    ]
    item = random.choice(items)
    inventory.append(item)
    print(f"Du hast {item} gefunden!")
    
    if item == "Magische Rüstung":
        has_magic = True
    elif item == "Flammenschwert":
        has_sword = True
    elif item == "Drachenschild":
        has_shield = True
    elif item == "Heiltrank":
        has_shield = True
    elif item == "Goldene Pfeile":
        has_shield = True
    elif item == "Geheimnisvolle Schriftrolle":
        has_magic = True
    elif item == "Lichtamulett":
        has_magic = True
    elif item == "Zauberstab":
        has_sword = True
    elif item == "Diamantdolch":
        has_shield = True
    elif item == "Schattenmantel":
        has_armor = True
    elif item == "Runenstein":
        has_magic = True
    elif item == "Eiszauber":
        has_sword = True
    elif item == "Heilige Kette":
        has_magic = True
    elif item == "Verstärkter Helm":
        has_sword = True
    elif item == "Blitzspeer":
        has_sword = True
    elif item == "Tarnumhang":
        has_armor = True
    elif item == "Dämonenring":
        has_sword = True
    elif item == "Verzauberte Schuhe":
        has_magic = True
    elif item == "Feueramulette":
        has_sword = True
    elif item == "Schutzring":
        has_armor = True
    elif item == "Villa":
        is_poor = False
        is_rich = True
        
    time.sleep(1)
    show_status()
    visit_shop()

def visit_shop():
    print("Du entdeckst einen Händler auf deiner Reise.")
    time.sleep(1)
    shop_items = {
        "Stärketrank": 30,
        "Feuerschwert": 50,
        "Mystischer Schild": 40,
        "Heiltrank": 20
    }
    print("Der Händler bietet folgende Gegenstände an:")
    for item, price in shop_items.items():
        print(f"{item}: {price} Gold")
    
    choice = input("Möchtest du etwas kaufen? (ja/nein): ").strip().lower()
    if choice == "ja":
        item_choice = input("Welches Item möchtest du kaufen?: ").strip()
        if item_choice in shop_items and konten["gold"] >= shop_items[item_choice]:
            konten["gold"] -= shop_items[item_choice]
            inventory.append(item_choice)
            print(f"Du hast {item_choice} gekauft! Verbleibendes Gold: {konten['gold']}")
        else:
            print("Du hast entweder nicht genug Gold oder das Item existiert nicht.")
            visit_shop()

def visit_markler():
    print("Du entdeckst einen Markler an seinem Stand.")
    time.sleep(1)
    shop_items = {
        "Villa": 1010,
        "Haus": 500,
        "halbes Haus": 250,
        "Mülltonne": 5
    }
    print("Der Händler bietet folgende Gegenstände an:")
    for item, price in shop_items.items():
        print(f"{item}: {price} Gold")
    
    choice = input("Möchtest du etwas kaufen? (ja/nein): ").strip().lower()
    if choice == "ja":
        item_choice = input("Welches Item möchtest du kaufen?: ").strip()
        if item_choice in shop_items and konten["gold"] >= shop_items[item_choice]:
            konten["gold"] -= shop_items[item_choice]
            inventory.append(item_choice)
            print(f"Du hast {item_choice} gekauft! Verbleibendes Gold: {konten['gold']}")
        else:
            print("Du hast entweder nicht genug Gold oder das Item existiert nicht.")
            visit_markler()

def visit_händler():
    print("Du bist zu dem Händler gegangen. Gute Wahl, dieser bietet dir alles besonders günstig an.")
    time.sleep(1)
    shop_items = {
        "Stärketrank": 20,
        "Feuerschwert": 25,
        "Mystischer Schild": 30,
        "Heiltrank": 10
    }
    print("Der Händler bietet folgende Gegenstände an:")
    for item, price in shop_items.items():
        print(f"{item}: {price} Gold")
    
    choice = input("Möchtest du etwas kaufen? (ja/nein): ").strip().lower()
    if choice == "ja":
        item_choice = input("Welches Item möchtest du kaufen?: ").strip()
        if item_choice in shop_items and konten["gold"] >= shop_items[item_choice]:
            konten["gold"] -= shop_items[item_choice]
            inventory.append(item_choice)
            print(f"Du hast {item_choice} gekauft! Verbleibendes Gold: {konten['gold']}")
        else:
            print("Du hast entweder nicht genug Gold oder das Item existiert nicht.")
            visit_händler()
    show_status()

def River():
    acquire_item()
    show_status()
    print("Dort ist ein Fluss mit einer Brücke aber auch ein Weg am Fluss entlang.")
    time.sleep(1)
    decision = input("Wohin gehst du: am Fluss entlang oder über die Brücke? (brücke/fluss): ").strip().lower()
    
    if decision == "brücke":
        find_gold()
        show_status()
        visit_shop()
        Markt()
    elif decision == "fluss":
        print("Oje, es regnet und das Wasser steigt. Du kannst nicht schwimmen und ertrinkst somit.")
        game_end()
    else:
        print("Ungültige Eingabe. Bitte gib 'brücke' oder 'fluss' ein.")
        River()

def Höhle():
    find_gold()
    acquire_item()
    proceed_to_dragon()

def Markt():
    show_status()
    print("Das hier sieht doch schon besser aus.")
    print("Viele glückliche Menschen und viele Stände. Du bist hier auf einem Markt gelandet.")
    time.sleep(1)
    decision = input("Ein Händler ruft nach dir, er will dir was verkaufen. Wirst du dies tun oder wirst du ihn ignorieren? (händler/ignorieren): ").strip().lower()
    if decision == "händler":
        show_status()
        visit_händler()
        Marktplatz()
        visit_markler()
    elif decision == "ignorieren":
        print("Wahrscheinlich die bessere Entscheidung. Der wollte dich sicher nur abziehen.")
        Marktplatz()
        visit_markler()
    else:
        print("Ungültige Eingabe. Bitte gib 'händler' oder 'ignorieren' ein.")
        Markt()

def Marktplatz():
    print("Dieser Marktplatz ist sehr zentral, von hier aus kommst du überall hin.")
    decision = input("Wohin möchtest du gehen (fluss/kreuzung/höhle): ").strip().lower()
    if decision == "fluss":
        River()
    elif decision == "kreuzung":
        Kreuzung()
    elif decision == "höhle":
        Höhle()
    else:
        print("Falsche Eingabe")
        Marktplatz()

def proceed_to_dragon():
    print("Nachdem du dich ausgerüstet hast, stehst du vor einer dunklen Höhle. Du spürst eine unheimliche Präsenz...")
    time.sleep(2)
    input("Drücke ENTER, um die Höhle zu betreten und dich dem Drachen zu stellen...")
    dragon_encounter()

def dragon_encounter():
    global has_armor, has_sword, has_shield, has_magic, inventory, is_rich, is_poor
    print(f"Ein riesiger {name} erscheint plötzlich!")
    print(r"                 ___====-_  _-====___               ")
    print(r"           _--^^^#####//      \#####^^^--_         ")
    print(r"        _-^##########// (    ) \##########^-_      ")
    print(r"       -############//  |\^^/|  \############-     ")
    print(r"     _/############//   (⚈::⚈)   \############\_.  ")
    print(r"    /#############((     \\//     ))#############\  ")
    print(r"   -###############\     (oo)    //###############- ")
    print(r"  -#################\  / \"\" \ //#################-")
    time.sleep(1)
    if name == "Zwergdrache":
        print("Oje der Zwergdrache. ist einfach zu mächtig.")
        print("Er ist so klein das du ihn nicht sehen kannst.")
        print("Diesen Kampf hat du verloren")
        time.sleep(2)
        print(r"                 ___====-_  _-====___                ")
        print(r"           _--^^^#####//      \#####^^^--_          ")
        print(r"        _-^##########// (    ) \##########^-_       ")             
        print(r"       -############//  |\^^/|  \############-      ")
        print(r"     _/############//   (⚈::⚈)   \############\_.   ")
        print(r"    /#############((     \\//     ))#############\   ")
        print(r"   -###############\     (oo)    //###############-  ")
        print(r"  -#################\  / \"\" \ //#################-")
        game_end()
    else:
        print("Du musst eine zufällige Zahl zwischen 1 und 10 erraten, um ihn zu besiegen.")
        show_status()
        dragon_number = random.randint(1, 10)
    
    try:
        guess = int(input("Welche Zahl wählst du? (1-10): "))
        if guess == dragon_number or has_sword:
            print("Du hast den Drachen besiegt und den Schatz gewonnen!")
            print(r"                 ___====-_  _-====___")
            print(r"           _--^^^#####//      \#####^^^--_")
            print(r"        _-^##########// (    ) \##########^-_")
            print(r"       -############//  |\^^/|  \############-")
            print(r"     _/############//   (X::X)   \############\_")
            print(r"    /#############((     \\//     ))#############\ ")
            print(r"   -###############\     (oo)    //###############-")
            print(r"  -#################\  / \"\" \ //#################-")
            print("Du erhältst 1000 Goldmünzen als Belohnung!")
            konten["gold"] += 1000
            Marktplatz()
        else:
            if has_armor:
                print("Deine Rüstung hat den ersten Angriff abgewehrt! Du hast eine zweite Chance.")
                has_armor = False
                dragon_encounter()
            elif has_shield:
                print("Dein Schild blockt den Angriff, aber zerbricht dabei! Du hast eine zweite Chance.")
                has_shield = False
                dragon_encounter()
            elif has_magic:
                print("Dein magischer Gegenstand hat dich geschützt.")
                has_magic = False
                dragon_encounter()
            else:
                print(r"                 ___====-_  _-====___                ")
                print(r"           _--^^^#####//      \#####^^^--_          ")
                print(r"        _-^##########// (    ) \##########^-_       ")
                print(r"       -############//  |\^^/|  \############-      ")
                print(r"     _/############//   (⚈::⚈)   \############\_.   ")
                print(r"    /#############((     \\//     ))#############\   ")
                print(r"   -###############\     (oo)    //###############-  ")
                print(r"  -#################\  / \"\" \ //#################- ")
                print(f"Der Drache besiegt dich! Die richtige Zahl war {dragon_number}.")
                game_end()
    except ValueError:
        print("Ungültige Eingabe. Bitte gib eine Zahl zwischen 1 und 10 ein.")
        dragon_encounter()
def game_end():
    if is_rich:
        print("Du bist reich gestorben.")
    elif is_poor:
        print("Jetzt wurdest du aus deiner Armut erlöst.")
    print("Spiel beendet.")
    exit()

def dragon_name():
    return random.choice([
        "Zwergdrache", "Feuerdrache", "Eisdrache", "dumme Drache", "gamer Drache", "coder Drache",
         "maßiver Drachenlord mit Lowtaperfait und riesen gyat "  
    ])
start_game()
```
